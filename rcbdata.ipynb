```python
# M Chinnaswamy Stadium (city=Bangalore)
Bangalore=matchinfodata[matchinfodata["city"]=="Bangalore"]
Bangalore
```

```python
# M Chinnaswamy Stadium (city=Bangalore) (winner=Bangalore)
Bangalorewin=Bangalore[Bangalore["winner"]=="Royal Challengers Bangalore"]
Bangalorewin
```

```python
# M Chinnaswamy Stadium (city=Bangalore) (winner=Bangalore) total win count
Bangalorewincount=len(Bangalorewin)
Bangalorewincount
```

```python
# M Chinnaswamy Stadium (city=Bangalore) (lost=Bangalore) 
Bangalorelost=Bangalore[Bangalore['winner']!="Royal Challengers Bangalore"]
Bangalorelost
```

```python
# M Chinnaswamy Stadium (city=Bangalore) (lost=Bangalore) totallost count
Bangalorelostcount=len(Bangalorelost)
Bangalorelostcount
```

```python
#summary of win and lost
print(" Royal Challengers Bangalore Wins in M Chinnaswamy Stadium:", Bangalorewincount)
print(" Royal Challengers Bangalore lost in M Chinnaswamy Stadium:", Bangalorelostcount)

```

```python
#Toss Decision field Win

tossfieldwin = Bangalore[(Bangalore['toss_decision'] == 'field') & (Bangalore['winner'] == 'Royal Challengers Bangalore')]
tossfieldwin
```

```python
#Count of Toss Decision field Win

tossfieldwincount = len(tossfieldwin)
tossfieldwincount
```

```python
#Toss Decision field Loss
tossfieldlost = Bangalore[(Bangalore['toss_decision'] == 'field') & (Bangalore['winner'] != 'Royal Challengers Bangalore')]
tossfieldlost
```

```python
#count of Toss Decision field Loss

tossfieldlostcount = len(tossfieldlost)
tossfieldlostcount
```

```python
#Toss Decision bat Win

tossbatwin = Bangalore[(Bangalore['toss_decision'] == 'bat') & (Bangalore['winner'] == 'Royal Challengers Bangalore')]
tossbatwin
```

```python

#Count of Toss Decision field Win

tossbatwincount = len(tossbatwin)
tossbatwincount
```

```python
#Toss Decision bat lost

tossbatlost = Bangalore[(Bangalore['toss_decision'] == 'bat') & (Bangalore['winner'] != 'Royal Challengers Bangalore')]
tossbatlost
```

```python
#count of Toss Decision field Loss

tossbatlostcount = len(tossbatlost)
tossbatlostcount
```

```python
#summary
print("Toss Decision: Bat | Match Win Count:", tossbatwincount)
print("Toss Decision: Bat | Match Lost Count:", tossbatlostcount)
print("Toss Decision: Field | Match Win Count:", tossfieldwincount)
print("Toss Decision: Field | Match Lost Count:", tossfieldlostcount)
print("Total Match RCB Win Count:", Bangalorewincount)
print("CTotal Match RCB Lost Count:", Bangalorelostcount)
```

```python
#Average number of matches won by RCB in Bangalore

total_matches = len(Bangalore)
total_wins = len(Bangalore[Bangalore['winner'] == 'Royal Challengers Bangalore'])

average_win = total_wins / total_matches

print("Average number of matches won by RCB  in Bangalore:", average_win)
```

```python
#Average number of matches LOST by RCB in Bangalore

total_matches = len(Bangalore)
total_lost = len(Bangalore[Bangalore['winner']!= 'Royal Challengers Bangalore'])

average_lost = total_lost / total_matches

print("Average number of matches lost by RCB  in Bangalore:", average_lost)
```
